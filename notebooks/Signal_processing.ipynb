```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

## Digital Low and Bandpass filters

### finite Impulse Response Filter

```python
def filter(taps,cut_off,rate,plot=False,pass_zero=True):
    """
        Function that computes the impulse response of the filter
    
    """
    h = sp.signal.firwin(taps, cut_off,pass_zero=pass_zero,width=5, nyq=rate/2)
    if plot:
        plt.plot(h,'.-')
        plt.title('Impulse respose of the low pass FIR digital filter')
        plt.show()
    return h
```

```python
N=100000
f0=10000 #1KHz, signal  frequency 
h=filter(101,[1000,15000],44000,pass_zero='bandpass',plot=True);
```

```python
H=np.fft.fft(h)
plt.plot(np.abs(H))
```
